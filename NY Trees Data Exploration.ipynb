```python
ny_trees['health'].fillna("Good", inplace=True)
ny_trees['spc_latin'].fillna("Unknown", inplace=True)
ny_trees['steward'].fillna("None", inplace=True)
ny_trees['sidewalk'].fillna("NoDamage", inplace=True)
ny_trees['problems'].fillna("None", inplace=True)
```

```python
large_trees = ny_trees[(ny_trees.stump_diam > 60) | (ny_trees.tree_dbh > 60)]
large_trees.shape
```

```python
ny_trees = ny_trees[(ny_trees.stump_diam <= 60) & (ny_trees.tree_dbh <= 60)]
ny_trees.shape
```

```python
ny_trees_alive = ny_trees[ny_trees.status == 'Alive']
ny_trees_dead_or_stump = ny_trees[(ny_trees.status == 'Dead') & (ny_trees.status == 'Stump')]
```

```python
stats_alive = ny_trees_alive.groupby('spc_latin')['tree_dbh'].describe().reset_index()[['spc_latin', '25%', '75%']]
stats_alive
```
